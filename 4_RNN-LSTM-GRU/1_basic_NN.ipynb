```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_regression()
X.shape, y.shape
```

```python
def linear_regression(X, y, lr = 0.001):
    w = np.random.random([300,100])
    b = np.random.random(100)
    m = b.shape[0]

    for i in range(1000):
        print('Epoch', i)

        # forward pass
        y_pred = w@X + b
        loss = (y_pred - y)**2
        
        # loss calculation
        loss = loss.mean()
        print(loss)

        # backward pass
        dJ_dw = (1/m)*(y_pred - y)@X
        dJ_db = (1/m)*(y_pred - y).sum()
        print(dJ_db.shape, dJ_dw.shape)

        w = w - dJ_dw *lr
        b = b - dJ_db *lr
        
```

```python
linear_regression(X, y)
```
